```python
import pandas as pd
```

```python
df = pd.read_csv('course_data.csv')
df
```

```python
def fee_calculator(duration):
    monthly_fee = 2000
    return monthly_fee * duration
```

```python
df['num_months']=df['Duration'].str[0]
df['num_months']=df['num_months'].astype(int)
```

```python
df
```

```python
df['total_fee'] = df['num_months'].apply(fee_calculator)
df['total_fee']
```
