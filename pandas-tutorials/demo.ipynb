Introduction to Pandas
Pandas is used for Python Data Analytics. It performs the operation of Data Cleaning and Transformation.

It is used for representing the data in tabular form.

It is far more efficient than Excel.

```python
import pandas as pd
import numpy as np
```

```python
data = np.array([100,200,300,400,500])
print(data)

ds = pd.Series(data)
```

```python
print(ds)
```

```python
cities = np.array(['Solapur','Pune','Chennai','Mumbai','Hyderabad','Bengaluru'])
print(cities)
```

```python
cs = pd.Series(cities)
print(cs)
```

```python
students = {
    'name':['Nihal','Gufran','Ahmed','Faiz','Sadik','Onkar'],
    'city':['Solapur','Mumbai','Lahor','Chennai','Kabul','Delhi'],
    'mark':[75,88,94,69,87,64]
}
print(students)
```

```python
df = pd.DataFrame(students)
print(df)
```
