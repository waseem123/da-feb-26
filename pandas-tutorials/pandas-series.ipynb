A Pandas Series is a 1-D array-like object. It can hold multiple datatypes like int, float, str,etc. It can operate like a Single Column.

```python
import numpy as np
import pandas as pd
```

```python
data = [100,200,300,400,500]
```

```python
s = pd.Series(data)
print(s)
```

```python
mylist = [100,'Solapur',45.89]
s = pd.Series(mylist)
print(s)
```

```python
mydict = {
    'empId':101,
    'empName':'Sam',
    'empSalary':55000
}

s = pd.Series(mydict)
print(s)
```

Working with loc() and iloc() function

```python
data = ['Solapur','Mumbai','Pune','Delhi','Chennai']
s = pd.Series(data)
print(s)
```

```python
print(s[0])
print(s.loc(0))
print(s.loc[0])
```

```python
mydict = {
    'empId':101,
    'empName':'Sam',
    'empSalary':55000
}

s = pd.Series(mydict)
print(s)
```

```python
print(s.loc['empId'])
print(s.loc['empName'])
```

```python
s.iloc[1]
print(s.iloc[2])
```

```python
print(s.iloc[2]+500)
```

```python
data = [100,200,300,400,500]
s = pd.Series(data)
x = s>200
print(x)
print(x.loc[2])
```

```python
y = pd.Series(s[s>200],index=[0,1,2,3,4])
print(y)
print(type(y))
```
