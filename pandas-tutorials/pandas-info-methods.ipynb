```python
def getAbbrevations(course_name):
    course_str = str(course_name)
    courselist = course_str.split(' ')
    abbvrt = ''
    for j in courselist:
        abbvrt += j[0]
    return abbvrt

# x = getAbbrevations(domainwise_courses['Course'])
# print(x)
```

```python
# domainwise_courses['abvr'] = ['a','b','c','d','e','f','g']
domainwise_courses['abvr'] = domainwise_courses['Course'].apply(getAbbrevations)
domainwise_courses
```
