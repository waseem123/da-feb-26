```python
import pandas as pd
```

```python
dataset = {
    'dept':['IT','Sales','IT','Marketing','Sales','Purchase','Sales','HR','HR','Marketing'],
    'salary':[55000,25000,80000,20000,45000,30000,40000,45000,45000,20000]
}
```

```python
df = pd.DataFrame(dataset)
df
```

```python
dept_df=df.groupby('dept')
dept_df
```

```python
total_salary=dept_df.sum()
total_salary
```
