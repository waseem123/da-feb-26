```python
student_marks = {
    'names' : ['Stud1','Stud2','Stud3','Stud3','Stud5','Stud3'],
    'c_marks' : [78,75,98,95,96,88],
    'cpp_marks' : [88,86,79,65,50,67],
    'java_marks': [77,81,96,93,92,75]
}
```

```python
marks_table = pd.DataFrame(student_marks)
print(marks_table)
```

```python
print(marks_table.iloc[2,2])
```

```python
row = marks_table.iloc[4,1:4]
print('Total marks - ',row.sum())
print(f'Percentage - {row.mean():.2f}')
```

```python
marks_table = pd.DataFrame(student_marks)
marks_table.set_index('names',inplace=True)
print(marks_table)
```

```python
print(marks_table.loc['Stud5'])
```

```python
print(marks_table.loc['Stud3','cpp_marks'])
```

```python
print(marks_table.iloc[0:3])
```

```python
print(marks_table.iloc[[0,3,4],[0,1]])
```

```python
print(marks_table.sum(axis=0))
```

```python
print(marks_table.iloc[0:,0])
```

```python
print(marks_table.sum(axis=1))
```

```python
total = marks_table.sum(axis=1)
percentage = marks_table.mean(axis=1)
```

```python
marks_table['total'] = total
marks_table['percentage'] = percentage.round(2)
marks_table
```
