Feature Engineering: Create a new column named Years_of_Service based on how many years they have been at the company relative to today.

```python
today = pd.to_datetime('today')
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
# 2. Calculate the total difference in days
total_days = (today - df['Date']).dt.days
total_years = total_days//365
total_months = (total_days%365)//30.44    
df['Years_of_Service'] = (total_years.astype(int).astype(str)+' Yr '+total_months.astype(int).astype(str)+' Mt')
df
```
